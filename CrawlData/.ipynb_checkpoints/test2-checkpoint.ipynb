```python
import requests
import json
import datetime as dt
import pandas as pd
from bs4 import BeautifulSoup
```

```python
def cal_std_day(befor_day):   
    x = dt.datetime.now() - dt.timedelta(befor_day)
    year = x.year
    month = x.month if x.month >= 10 else '0'+ str(x.month)
    day = x.day if x.day >= 10 else '0'+ str(x.day)  
    return str(year)+ '-' +str(month)+ '-' +str(day)
```

```python
def get_bs_obj(url, headers = None, params = None):
    result = requests.get(url, headers = headers, params = params)
    soup = BeautifulSoup(result.content, 'html.parser')
    return soup
```

```python
base_url=['302','301']
data = []
cols = ['종목명', '코스피/코스닥' ,'현재가','전일비','등락률','매수호가','매도호가','거래량','거래대금','전일거래량','업종명']
for i in base_url:
    url = 'https://finance.naver.com/sise/sise_group_detail.naver?type=upjong&no={}'.format(i)
    response = requests.get(url)
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    trs = soup.findAll('table',{'class':'type_5'})[0].find('tbody').findAll('tr')[:-2]
    for tr in trs :
        rows = []
        # 종목명
        if '*' in tr.findAll('td')[0].text:
            rows.append(tr.findAll('td')[0].text.replace('*','').replace(' ',''))
            rows.append('코스닥')
        else :
            rows.append(tr.findAll('td')[0].text.replace('*','').replace(' ',''))
            rows.append('코스피')  
        # 현재가
        rows.append(tr.findAll('td')[1].text.replace(',','').replace(' ',''))
        # 전일비
        rows.append(tr.findAll('td')[2].text.strip())
        # 등락률
        rows.append(tr.findAll('td')[3].text.strip())
        # 매수호가
        rows.append(tr.findAll('td')[4].text.replace(',','').replace(' ',''))
        # 매도호가
        rows.append(tr.findAll('td')[5].text.replace(',','').replace(' ',''))
        # 거래량
        rows.append(tr.findAll('td')[6].text.replace(',','').replace(' ',''))
        # 거래대금
        rows.append(tr.findAll('td')[7].text.replace(',','').replace(' ',''))
        # 전일거래량
        rows.append(tr.findAll('td')[8].text.replace(',','').replace(' ',''))
        # 업종명
        rows.append(soup.find('div',{'id':'contentarea_left'}).find_all('td')[1].text.strip())
        tmp = dict(zip(cols, rows))    
        data.append(tmp) 

res = {
    'meta':{
        'desc':'업종별 종목',
        'cols':{
            '종목명':'종목명'
            ,'코스피/코스닥':'코스피/코스닥유무'
            ,'현재가':'현재가'
            ,'전일비':'전일비'
            ,'등락률':'등락률'
            ,'매수호가':'매수호가'
            ,'매도호가':'매도호가'
            ,'거래량':'거래량'
            ,'거래대금':'거래대금'
            ,'전일거래량':'전일거래량'
            ,'업종명':'업종명'
        },
        'std_day':cal_std_day(1)
    },
   'data':data
}
data
```
